```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\buset\Downloads\datset - Sheet1.csv")
```

```python
df
```

```python
# Display initial info
print("\nInitial Data:")
print(df.head())
```

```python
# Step 2: Clean missing/null values
df_clean = df.dropna(subset=["Region", "Product", "Quantity", "UnitPrice", "TotalSales"])
print("\nAfter Dropping Rows with Null Values:")
print(df_clean.isnull().sum())
```

```python
# Step 3: Sum of sales by region
sales_by_region = df_clean.groupby("Region")["TotalSales"].sum()
print("\nTotal Sales by Region:")
print(sales_by_region)
```

```python
# Step 4: Average sales per product
avg_sales_per_product = df_clean.groupby("Product")["TotalSales"].mean()
print("\nAverage Sales per Product:")
print(avg_sales_per_product)
```

```python
# Step 5: Highest and lowest selling products
total_sales_per_product = df_clean.groupby("Product")["TotalSales"].sum()
highest_selling = total_sales_per_product.idxmax()
lowest_selling = total_sales_per_product.idxmin()
```

```python
total_sales_per_product
```

```python
print("\nHighest Selling Product:", highest_selling)
print("Lowest Selling Product:", lowest_selling)
```

```python
# Step 6: NumPy statistics
print("\nNumPy Statistics:")
sales = df_clean["TotalSales"].values
print("Mean Total Sales:", np.mean(sales))
print("Median Total Sales:", np.median(sales))
print("Standard Deviation of Total Sales:", np.std(sales))
```
